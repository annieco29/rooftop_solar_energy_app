```python
lat = []
longi = []

from geopy.geocoders import Nominatim
for addr in df['address']:
    geolocator = Nominatim(user_agent="florida_industrial_sites")
    location = geolocator.geocode(addr, timeout=10000, language = 'en')
    try:
        #city.append(location.address.split(',')[-4])
        lat.append(location.latitude)
        longi.append(location.longitude)
    except:
        print(addr)
```

```python
len(lat)
```

```python
d = {'lat':lat,'long':longi}
```

```python
df_lat_long = pd.DataFrame(d)
```
