```python
from os import path
home = path.abspath(path.join("__file__",".."))
print(home)
import os
import csv

# fields = []
# rows = []

# for filename in os.listdir(home + '/tampa_weather_data_2017_2019/'):
#     print(filename)
#     #filename = pd.read_csv(os.path.join(home + '/tampa_weather_data_2017_2019/', filename)

import pandas as pd
import glob

path = home + '/florida_weather_data_by_county/' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    print(filename)
    df = pd.read_csv(filename, index_col=None, header=0)
    print(df.iloc[0]['Latitude'])
    lat = df.iloc[0]['Latitude']
    print(df.iloc[0]['Longitude'])
    long = df.iloc[0]['Longitude']
    df = df.iloc[1: , :]
    df.columns = df.iloc[0]
    df = df.drop(df.index[0])
    df = df[['Year','Month','Day','Hour','Minute','DHI','DNI','Dew Point','Surface Albedo','Wind Speed','Relative Humidity','Temperature','Pressure']]
    df['lat'] = lat
    df['long'] = long
    display(df.head(5))
    print(df.shape)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame['Year'] = pd.to_numeric(frame['Year'], errors = 'coerce')
frame['Month'] = pd.to_numeric(frame['Month'], errors = 'coerce')
frame['Day'] = pd.to_numeric(frame['Day'], errors = 'coerce')
frame['Hour'] = pd.to_numeric(frame['Hour'], errors = 'coerce')
frame['Minute'] = pd.to_numeric(frame['Minute'], errors = 'coerce')
frame['DHI'] = pd.to_numeric(frame['DHI'], errors = 'coerce')
frame['DNI'] = pd.to_numeric(frame['DNI'], errors = 'coerce')
frame['Dew Point'] = pd.to_numeric(frame['Dew Point'], errors = 'coerce')
frame['Surface Albedo'] = pd.to_numeric(frame['Surface Albedo'], errors = 'coerce')
frame['Wind Speed'] = pd.to_numeric(frame['Wind Speed'], errors = 'coerce')
frame['Relative Humidity'] = pd.to_numeric(frame['Relative Humidity'], errors = 'coerce')
frame['Temperature'] = pd.to_numeric(frame['Temperature'], errors = 'coerce')
frame['Pressure'] = pd.to_numeric(frame['Pressure'], errors = 'coerce')
frame['lat'] = pd.to_numeric(frame['lat'], errors = 'coerce')
frame['long'] = pd.to_numeric(frame['long'], errors = 'coerce')
```

```python
frame_group_by_date = frame.groupby(['lat','long','Year','Month','Day']).mean()
```

```python
frame_group_by_date['date_time'] = pd.to_datetime(frame_group_by_date[['Year', 'Month', 'Day']])
```
